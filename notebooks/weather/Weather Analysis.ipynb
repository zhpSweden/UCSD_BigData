### NaN-tolerant averaging  ###
We compute the empirical covariance matrix in a way that tolerates NaN values.

<span style="color:red"> In the code below I remove all rows that have a nan in them. If you remve the command **M.dropna(...** then all rows are used. Can you get better results without removing the rows? </span>

```python
# demonstrating the use of the cell magic %%time, which measures the run-time of the cell.
M=Dout.loc[:,1:365].transpose()
M=M.dropna(axis=1)
(columns,rows)=shape(M)
Mean=mean(M, axis=1).values

print (columns,rows), shape(Mean)
C=np.zeros([columns,columns])   # Sum
N=np.zeros([columns,columns])   # Counter of non-nan entries
```

```python
%%time
for i in range(rows):
    if i % 1000==0: 
        print i
    row=M.iloc[:,i]-Mean;
    outer=np.outer(row,row)
    valid=isnan(outer)==False
    C[valid]=C[valid]+outer[valid]  # update C with the valid location in outer
    N[valid]=N[valid]+1
valid_outer=np.multiply(1-isnan(N),N>0)
cov=np.divide(C,N)
```

```python
U,D,V=np.linalg.svd(cov)
```

<span style="color:red">Question 1</span>

<span style="color:red">Can you reduce the reconstruction error (using a fixed number of eigenvectors) by splitting the stations according to region (for example country, state, latitudal range). Note that having a regions with very few readings defeats the purpose.

My scenario here is to partition the sample data into groups based on their station country code. According to the file 'ghcnd-countries.txt' located in the ftp site: http://www1.ncdc.noaa.gov/pub/data/ghcn/daily/, the first two characters of the station indicate the country it belongs to. For example, USC00054664 means that it is a US station. My idea here is to build a hashtable in which the key is the country code and the value is a pandas DataFrame that stores the weather cases belonging to that specific country. 

I refactored some code above and rewrote into different functions to avoid the messy thing that a change in a part leads to a series to painful changes. The function names are quite self-explanatory and you may understand its semantics based on the brief comments 

```python
from collections import defaultdict
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import math

# to get the mean of all cases, the optional parameter remove_nan means whether we just eliminate the case that contains any nan,
# the default setting is False, which means we keep the nan and thus we need to take more case in computing the empirical covariance
# matrix. 
def getMean(df, remove_nan=False):
	M = df.loc[:,1:365].transpose()
	if remove_nan: M = M.dropna(axis=1)
	Mean = np.mean(M, axis=1)
	return M, Mean

# as the name suggests, it is used to compute the eigenvectors
def compute_eigenvector(df, remove_nan=False):
	M, Mean = getMean(df, remove_nan)
	columns, rows = np.shape(M)
	C = np.zeros([columns,columns])
	N = np.zeros([columns,columns])

	for i in range(rows):
		row = M.iloc[:,i]-Mean
		outer = np.outer(row,row)
		valid = np.isnan(outer)==False
		C[valid] = C[valid]+outer[valid]  # update C with the valid location in outer
		N[valid] = N[valid]+1

	cov = np.divide(C,N)
	U,D,V = np.linalg.svd(cov)
	return U, Mean

# it invokes the function above and returns the reconstructed data that has been transposed
def reconstruct(df, k=50, remove_nan=False):
	U, Mean = compute_eigenvector(df, remove_nan)
	Eig = np.matrix(U[:,:k])
	matrix = np.matrix(df.ix[:,1:365])-Mean
	matrix[np.isnan(matrix)] = 0
	Prod = matrix*Eig
	Recon = Eig*Prod.transpose()+Mean[:,np.newaxis]
	return Recon

# it is used to get the row index of a certain case with a specific station name and year in a DataFrame
def getPos(df, station, year):
	indexlist = df.index.tolist()
	tmp = df[(df.station==station) & (df.year==year)]
	index = tmp.index[0]
	pos = indexlist.index(index)
	return pos

# used to plot the original case against the reconstructed case of a preordained station name and year
def plot_recon_specific(station, year, df, recon):
	i = getPos(df, station, year)
	plt.figure()
	recon = [elem[0] for elem in recon[:,i].tolist()]
	plt.plot(recon)
	plt.plot(list(df.ix[i,1:365]))
	plt.title(station+' / '+str(year))
	plt.xlim([0,365])
	plt.show()

# the same semantics as the one the teather provides
def plot_recon_subplot(df, recon, selection, rows=2,columns=7):
	plt.figure(figsize=(columns*3,rows*3),dpi=300)
	j=0
	for i in selection:
		plt.subplot(rows,columns,j)
		j += 1
		if j>=rows*columns: break
		tmp = [elem[0] for elem in recon[:,i].tolist()]
		plt.plot(tmp)
		plt.plot(list(df.ix[i,1:365]))
		plt.title(df.ix[i,'station']+' / '+str(df.ix[i,'year']))
		plt.xlim([0,365])
	plt.show()

# get the squared error of a reconstruced case with a specific station name and year
def error_specific(station, year, df, recon):
	pos = getPos(df, station, year)
	df = df.ix[:,1:365]
	diff = df.T-recon[:,pos]
	diff = np.multiply(diff, diff)
	return list(np.sum(diff, axis=0)/diff.shape[0])

# get a list of squared error of all the cases stored in the dataframe you input
def error(df, recon, selection=False):
	if not selection:
		selection = range(df.shape[0])
	df = df.ix[selection,1:365]
	recon = recon[:,selection]
	diff = df.T-recon
	diff = np.multiply(diff, diff)
	index = np.isnan(diff)
	col_num = index.shape[0]
	index = np.sum(index, axis=0)
	sumValue = np.sum(diff, axis=0)
	tmp = list(np.divide(sumValue, col_num-index))
	res = []
	for elem in tmp:
		if elem != np.inf: res.append(elem)
		else: res.append(0)
	return res
    
# get the root mean square of all the cases stored in the dataframe you input
def rms(df, recon, selection=False):
	e = error(df, recon, selection)
	index = np.isnan(e)
	for i, elem in enumerate(index):
		if elem:
			e[i] = 0
	return math.sqrt(sum(e)*1.0/(len(e)-sum(index)))
```

After refactoring the code, next I will build the hashtable. 'hashmap_country' is the hashtable I built. You can use hashmap_country['country_code'] to fetch the dataframe of the country

```python
Djoined = Djoined[['station','year','Mean','Std','latitude','longitude','elevation','AbsLatitude']+range(1,366)]

hashmap = defaultdict(list)
for i in range(Djoined.shape[0]):
    name = Djoined.at[i,'station'][0:2]
    hashmap[name].append(i)

hashmap_country = {}
for k, v in hashmap.items():
    value = Djoined.ix[v]
    value.index = range(value.shape[0])
    hashmap_country[k] = value
```

The RMS we define here is to first get the mean squared error of every case and then get the square root of the average of all the MSEs that we have. Next we define a function that is used to get the RMS of all the reconstructed cases that correspond to a specific country. It also plots the first 14 reconstructed cases against original ones. You may change the rows and columns setting in the plot function to get different number of plots. We first try the United States and see its RMS. 

We see that the RMS is 3.47163166398.

```python
def resultBasedOnCountry(code):
	df = hashmap_country[code]
	recon = reconstruct(df)
	num = recon.shape[1]
	plot_recon_subplot(df, recon, range(num))
	print rms(df, recon)

# get US error
resultBasedOnCountry('US')
```

Next we need to compare the result of the 'splitting by country' method with the original one that computes the eigenvectors based on the entire dataset as opposed to calculating different eigenvector bases for different countries. 

First we get the recontructed cases of the entire dataset.

```python
recon_origin = reconstruct(Djoined)
```

Then we pull out the cases that correspond to US and plot the first 14 cases and also get its RMS. We see that the reconstruction error is 3.23672858619, which means the 'splitting by country' method does reduce the reconstruction error for US. You may be doubtful of the seemingly tiny error difference, which actually is not as negligible as it appears. Since the recontruction error we define here is the RMS, which means first get the sum of all the squared errors of all the cases and then get the square of its mean value. With the knowledge of the underlying computing process, we can understand that a small difference in RMS actually means a significant reduction in the reconstruction error. 

```python
def originErrorBasedOnCountry(code):
	index = Djoined.station.apply(lambda x: x[:2] == code)
	data_country = Djoined[index]
	data_country.index = range(data_country.shape[0])
	tmp_index = []
	for i, elem in enumerate(index):
		if elem:
			tmp_index.append(i)
	recon_matrix = recon_origin[:, tmp_index]
	num = np.shape(recon_matrix)[1]
	plot_recon_subplot(data_country, recon_matrix, range(num))
	origin_error = error(data_country, recon_matrix)
	print rms(data_country, recon_matrix)
```

To increase the calculating efficiency, we use the datasets with nan rows dropped. For the original case, we use k = 50 and get rms equal to 3.71963615601. Now we try k=60, 70, 80, 90 and 100 respectively and see how the RMS decreases as we increase the number.

```python
RMS = []
num = [50,60,70,80,90,100]
for elem in num:
    recon = reconstruct(Dout, k=elem, remove_nan=True)
    RMS.append(rms(Dout, recon))
print RMS
```
